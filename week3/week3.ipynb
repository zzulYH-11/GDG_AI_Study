```python
# 농어 데이터 다시 입력
import numpy as np

perch_length = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
# 훈련/테스트 세트 나누고 재구성
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42)

train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
# K-최근접 이웃 회귀 모델 훈련 (n_neighbors=3)
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(train_input, train_target)
```

```python
# 50cm 농어 무게 예측
print(knr.predict([[50]]))
```

```python
# 50cm 농어 예측 및 이웃 시각화
import matplotlib.pyplot as plt

distances, indexes = knr.kneighbors([[50]])

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
plt.scatter(50, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 50cm 농어 이웃들의 평균 무게 출력
print(np.mean(train_target[indexes]))
```

```python
# 100cm 농어 무게 예측
print(knr.predict([[100]]))
```

```python
# 100cm 농어 예측 및 이웃 시각화
distances, indexes = knr.kneighbors([[100]])

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
plt.scatter(100, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 선형 회귀 모델 임포트
from sklearn.linear_model import LinearRegression
```

```python
# 선형 회귀 모델 훈련
lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
# 선형 회귀 모델로 50cm 농어 무게 예측
print(lr.predict([[50]]))
```

```python
# 선형 회귀 모델의 계수와 절편 출력
print(lr.coef_, lr.intercept_)
```

```python
# 선형 회귀선과 예측점 시각화
plt.scatter(train_input, train_target)
plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])
plt.scatter(50, 1241.8, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 선형 회귀 모델 훈련 및 테스트 세트 점수 평가
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

```python
# 2차항을 추가한 특성 생성
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
```

```python
# 2차항이 추가된 훈련/테스트 데이터 형태 출력
print(train_poly.shape, test_poly.shape)
```

```python
# 다항 회귀 모델 훈련 및 50cm 농어 무게 예측
lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[50**2, 50]]))
```

```python
# 다항 회귀 모델의 계수와 절편 출력
print(lr.coef_, lr.intercept_)
```

```python
# 다항 회귀선과 예측점 시각화
point = np.arange(15, 50)
plt.scatter(train_input, train_target)
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05)
plt.scatter([50], [1574], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 다항 회귀 모델 훈련 및 테스트 세트 점수 평가
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

```python
# CSV 데이터 로드 및 NumPy 배열로 변환
import pandas as pd

df = pd.read_csv('https://bit.ly/perch_csv_data')
perch_full = df.to_numpy()
print(perch_full)
```

```python
# 농어 무게 데이터 입력
import numpy as np

perch_weight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
# 전체 농어 데이터를 훈련/테스트 세트로 분할
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state=42)
```

```python
# PolynomialFeatures 변환 예시
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
poly.fit([[2, 3]])
print(poly.transform([[2, 3]]))
```

```python
# bias 항을 포함하지 않는 PolynomialFeatures 변환 예시
poly = PolynomialFeatures(include_bias=False)
poly.fit([[2, 3]])
print(poly.transform([[2, 3]]))
```

```python
# 훈련 데이터에 다항 특성 적용
poly = PolynomialFeatures(include_bias=False)

poly.fit(train_input)
train_poly = poly.transform(train_input)

print(train_poly.shape)
```

```python
# 다항 특성의 이름 확인
poly.get_feature_names_out()
```

```python
# 테스트 데이터에 다항 특성 적용
test_poly = poly.transform(test_input)
```

```python
# 선형 회귀 모델 훈련 및 훈련 세트 점수 평가 (다항 특성)
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
```

```python
# 선형 회귀 모델 테스트 세트 점수 평가 (다항 특성)
print(lr.score(test_poly, test_target))
```

```python
# 5차 다항 특성 생성
poly = PolynomialFeatures(degree=5, include_bias=False)

poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
```

```python
# 5차 다항 특성 훈련 데이터 형태 출력
print(train_poly.shape)
```

```python
# 5차 다항 특성으로 선형 회귀 모델 훈련 및 점수 평가
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

```python
# 특성 스케일링
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
# 릿지 회귀 모델 훈련 및 점수 평가
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

```python
# 라쏘 회귀 모델 훈련 및 점수 평가
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# degree=15로 설정하여 다항 특성 생성
poly = PolynomialFeatures(degree=15, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)

# 다중 회귀 모델 학습
lr = LinearRegression()
lr.fit(train_poly, train_target)
```

```python
print('Train Score (degree=15):', lr.score(train_poly, train_target))
print('Test Score (degree=15):', lr.score(test_poly, test_target))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge

# StandardScaler로 데이터 스케일링
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)

# Ridge 규제 적용 (기본 alpha=1)
ridge = Ridge()
ridge.fit(train_scaled, train_target)
```

```python
print('Ridge Train Score:', ridge.score(train_scaled, train_target))
print('Ridge Test Score:', ridge.score(test_scaled, test_target))
```
